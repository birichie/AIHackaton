```python
import pandas as pd
```

```python
## Loading datasets
agric_tbl = pd.read_csv("agric_tbl.csv", encoding= 'unicode_escape')

## This dataset has been previously preprocessed to map the products in the agricultural dataset
mapped_nutrient_tbl = pd.read_csv("base_tbl.csv")
```

```python
## Collecting World Emissions data
world_data = agric_tbl[agric_tbl.Area == 'World']
```

```python
world_data.head()
```

```python
## Columns that speaks about emissions
emissions_columns = ['Eggs, hen, in shell-Emissions (CO2eq)-kilotonnes',
                    'Meat, buffalo-Emissions (CO2eq)-kilotonnes',
                     'Meat, cattle-Emissions (CO2eq)-kilotonnes',
                     'Meat, chicken-Emissions (CO2eq)-kilotonnes',
                     'Meat, goat-Emissions (CO2eq)-kilotonnes',
                     'Meat, pig-Emissions (CO2eq)-kilotonnes',
                     'Meat, sheep-Emissions (CO2eq)-kilotonnes',
                     'Milk, whole fresh buffalo-Emissions (CO2eq)-kilotonnes',
                     'Milk, whole fresh camel-Emissions (CO2eq)-kilotonnes',
                     'Milk, whole fresh cow-Emissions (CO2eq)-kilotonnes',
                     'Milk, whole fresh goat-Emissions (CO2eq)-kilotonnes',
                     'Milk, whole fresh sheep-Emissions (CO2eq)-kilotonnes',
                     'Rice, paddy-Emissions (CO2eq)-kilotonnes']
```

```python
## Get their average emissions across years
avg_emmisions = pd.DataFrame(agric_tbl[agric_tbl.Area == 'World'][emissions_columns].mean(axis=0)
```

```python
## Data preparation
avg_emmisions = avg_emmisions.to_frame().reset_index()
avg_emmisions.columns = ['product','emission_co2eq_kt']
avg_emmisions['product_name'] = (avg_emmisions['product'].str.split('-', expand = True))[0]
avg_emmisions = avg_emmisions[['product_name','emission_co2eq_kt']]
```

```python
avg_emmisions.head()
```

## Preparing the Nutrients Table

```python
test = mapped_nutrient_tbl[['product','daily_nutrient','product_nutrient_amount']]
```

```python
pivot_tbl = (test.pivot(index=['product'],columns="daily_nutrient", values="product_nutrient_amount")
         .sort_index(level=[1,0])).reset_index()
```

```python
pivot_tbl.head()
```

```python
## Merge emissions to product nutrients table
final_tbl = pd.merge(avg_emmisions, pivot_tbl, left_on='product_name', right_on = 'product', how='inner')
```

```python
final_tbl.head()
```

## Creating a random nutritional requirement

```python
required_food = pivot_tbl[(pivot_tbl.calcium > 3)\
                          & (pivot_tbl.cholesterol < 90)\
                          & (pivot_tbl.iron > 0.4)
                         ]['product'].to_list()
```

```python
required_food
```

```python
## Gets products that meets our custom requirements
filter_product = final_tbl['product'].isin(required_food)
```

```python
data_for_combination = final_tbl[filter_product][['product_name','emission_co2eq_kt']]
data_for_combination
```

```python
## Extract pieces
product = data_for_combination.product_name.values.tolist()
emmis = data_for_combination.emission_co2eq_kt.values.tolist()
```

```python
## Function to calculation Co2 combinations
def powerSet(items):
    """
    Power set generator: get all possible combinations of a list’s elements

    Input:
        items is a list
    Output:
        returns 2**n combination lists one at a time using a generator 

    Reference: edx.org 6.00.2x Lecture 2 - Decision Trees and dynamic programming
    """

    N = len(items)
    # enumerate the 2**N possible combinations
    for i in range(2**N):
        combo = []
        for j in range(N):
            # test bit jth of integer i
            if (i >> j) % 2 == 1:
                combo.append(items[j])
        yield combo
```

## Extract Information

```python
## Make all combinations of product and get the total Co2, product combination and number of products combined
product_comb = pd.Series([ x for x in powerSet(product)])
emis_comb = pd.Series([sum(x) for x in powerSet(emmis)])
n_comb = pd.Series([len(x) for x in powerSet(product)])
```

```python
## Convert to a dataframe
final_combination = pd.concat([product_comb, emis_comb,n_comb], axis=1)
final_combination.columns = ['products','emission_co2eq_kt','n_combinations']
```
